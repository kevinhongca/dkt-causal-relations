```python
import re
regex_pattern = '(?:^|_)(?:' + '|'.join(map(re.escape, nodes)) + r')(?:_|$)'

#ensures 'skill_id' is a string
assist_df['skill_id'] = assist_df['skill_id'].astype(str)

#filter df
filtered_df = assist_df[assist_df['skill_id'].str.contains(regex_pattern, regex=True)]
```

Check # of Exercises

```python
nodes2 = filtered_df['skill_id'].unique()
```

```python
pd.reset_option('display.max_rows')
queried_df2 = assist_df[assist_df['skill_id'].isin(nodes2)]
```

```python
df_counts = queried_df2['skill_id'].value_counts()
df_dict = df_counts.to_dict()
```

```python
from collections import defaultdict

store = df_dict.keys()
store2 = set()
store3 = defaultdict(int)
temp = []

for item in store:
    if isinstance(item, int):
        # print(item)
        store2.add(str(item))
        store3[str(item)] = df_dict[item]
    else:
        temp = item.split("_")
        for item2 in temp:
            store2.add(item2)
            store3[item2] += df_dict[item]

store3 = dict(store3)
total_sum = sum(store3.values())
total_sum
len(store2)
```
